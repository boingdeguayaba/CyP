2. Hacer un programa que solicite 3 números enteros diferentes entre sí y que imprima cuál de los 3 es el mayor

```python
num1=int(input("dame un numero: "))
num2=int(input("dame un segundo numero diferente al primero: "))
num3=int(input("dame un tercer numero diferente a los anteriores: "))
if num1 > num2 and num1 > num3:
    print(f"el primer numero que registraste {num1} es mayor que los otros: {num2} y {num3}")
elif num2 > num1 and num1 > num3:
    print(f"el segundo numero que registraste {num2} es mayor que los otros: {num1} y {num3}")
elif num1 == num2 and num1 == num3:
    print(f"los tres numeros registrados son iguales {num1}, {num2} y {num3}")
else:
    print(f"el tercer numero que registraste {num3} es mayor que los otros: {num1} y {num2}")
```
