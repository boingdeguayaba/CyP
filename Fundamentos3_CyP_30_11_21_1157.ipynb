Repetición:
Este tipo de estructura nos permite automatizar la ejecucion repetitiva de codigo mientras ciertas condiciones se cumplan, tenemos 3 variantes de etsa estructura en todos los lenguajes de programación:
- Desde...hasta(estructura for)
- haz mientras (estructura while)
- haz y luego valida mientras (estructura Do-While)

# Estructura de control de repetición Haz- mientras
Una estructura de control haz-mientras mos permite ejecutar repedidamente mientras una condicion se siga cumpliendo
Sintaxis
```
while <condion>:
    <bloque de sentencias a ejecutar repetidamente>
<siguiente sentencia posterior a while>
```
La diferencia principal en comparacion a for es que con while no conoces el numero de ciclos a ejecutar, en for si se conoce.

como por ejemplo iniciamos con el siguiente enunciado

Hacer un programa que solicite un valor entero y calule repetidamente la mitas y se reasigne a la misma variable. El ciclo se debe de detener cuando el resultado es menor a 1
```
introduce un numero entero postivo: 17
8.5
4.25
2.12
1.06
0.53
total de interaciones = 5
```

````python
valor= int(input("introduce un numero entero psoitivo"))
contador = 0
if valor> 0:
    while valor> 1.0:
        valor=valor /2
        print(round(valor,2))
        contador +=1
else:
    print("el numero es negativo, intenta de nuevo")
print("total de iteraciones=" , contador)
````

escribe un programa que permita generar la tabla de multiplicar de un numero N, comenzando desde el 1

Si el usuario escribe un numero incorrecto, el programa no se ejecuta, en cambio se pregunta hasta que el dato ingresado sea correcto

````python
comprobar= True
while comprobar== True:
    n=int(input("ingrese un numero positivo: "))
    if n>0:
        print("es correcto")
        for i in range(10):
            print(n, "x", i+1 , "=" , n*(i+1))
            comprobar==False
    else:
        print("el numero ingresado no es correcto")
````

Escrine un programa que al recibir como dato un numero entero positivo N, calcule el resultado de la siguiente serie

1+(1/2)+(1/3)+(1/4)+...+(1/N)
si el usuario escribe un numero incorrecto, el programa no se ejecuta. En cambio se pregunta de nuevo por la informacion hasta que el dato sea correcto

````python
comprobar= True
while comprobar == True:
    n=int(input("ingrese un numero positivo: "))
    if n>0:
        comprobar= False
        resultado=0
        for i in range(1,n+1):
            resultado+=(1/i)
            print("el resultado de la serie es", resultado)
    else:
        print("el numero ingresado no es correcto, intente de nuevo")
````
