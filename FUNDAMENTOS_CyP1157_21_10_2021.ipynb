# Slicing (rebanado) de cadenas

Perminte seleccionar subcadenas con el operador
#### [ inicio : stop : incremento ]
el stop no se incluye, es por eso que debemos agregar una posici贸n extra
### Slicing y valores por defecto
inicio por defecto siempre sera 0
stop: tama帽o del arreglo (indexado positivo)
        0 ( en un indexado negativo)

incremento: es 1 (indexado positivo)

```python
 carrera="Ingenier铆a en computaci贸n"
print(carrera[0:10:1])
print(carrera[14:25:1])
# lo mismo pero con indexado negativo
print(carrera[-11:-1:1])
print(carrera[:10:])
print(carrera[14:25:])
print(carrera[-11::1])
#soluci贸n previa
print("-------")
print(carrera[-11:-1:1])
print(carrera[-11::1])
## ejercicio de incremento
print(carrera[0:10:2])
print(carrera[-1:-12:-1])
print(carrera[-1:-12:-2])
```

#Funcion print
###Se emplea para desplegar informacion en la salida estandar.
Normalmente en un monitor.
##formas de usar print
- el argumento por defecto de print (end)
- formtat()
- operador f""



```python
nombre= "manuel"
edad= 17
casado= False
estatura= 1.74
"""
nombre: Manuel
  edad: 18
  esta casado: false
  estatura: 1.74
  """
# 1.- con comas .- concatena haciendo un casting (conversion a string automaticamente
#agrega un espacio)
print(nombre, edad, casado, estatura)
print("nombre: " +"nombre +\n\tedad: "+ str(edad) + "\n\testa casado : "+ str(edad) + "\n\tcuanto mide:"+str(estatura)) 

# 2.- con el sombolo de "+"
#concatenacion en crudo
print(nombre+" " + str(edad) + " " + str(casado)+" " + str(estatura))

# 3.- dando formato a la salida con format()
print("nombre: {}\n\tedad: {}\n\testa casado: {}\n\testatura en metros:{}".format(nombre,edad,casado,estatura))
print("nombre: {3}\n\tedad: {2}\n\testa casado: {1}\n\testatura en metros:{0}".format(nombre,edad,casado,estatura))

# 4.- con el operador f" "
print("con el operador f\" \"")
print(f"nombre: {nombre}\n\tedad: {edad}\n\testa casado: {casado}\n\testatura en metros:{estatura}")

#como manejar el argumento end
print("\n como usar el argumento end")
print(f"Hola {nombre}" , end="")
print(", como estas?" , end="\n")
```
