# La palabra reservada None de python 
Esta palabra significa vacio, equivalente a la palabra reservasa void del lenguaje c, c++ y java

Se usa para indicar ya sea un return vacio o que se recive  como parametro un vacio

```python
def multiplicar(valor, veces):
    c=valor*veces
    return c
```

```python
res=multiplicar(5,5)
print(res)
```

```python
def multiplicar(valor, veces):
    if valor==None:
        c=-1
    else:
        c=valor*veces
    return c
```

```python
res=multiplicar(None,5)
print(res)
```

# Argumantos posicionales 
En python es posible determinar a que parametro estan dirigidos los argumentos, segun su posicion.

```python
def dividir(op1,op2):
    c=op1/op2
    return c
```

```python
num1=25
num2=5
res=dividir(num1,num2)
print(res)

res=dividir(num2,num1)
print(res)
```

# argumentos con palabras clave (keywords)
Es la posibilidad de indicarle a la funcion en tiempo de ejecucion a que parametro va dirigido un argumeto de entrada

Solo disponible en python

```python
res=dividir(op2=num2,op1=num1)
print(res)
```
